# 1. Building Knowledge Graph

## 1.1 Extract Data

```python
import csv
```

```python
def get_data_from_csv(filepath: str):
    data = []
    with open(filepath) as file:
        next(file)
        rows = csv.reader(file, delimiter=",")
        for row in rows:
            num,kanji,radicals,meanings,theme,subtheme = row
            radicals = radicals.split(':')
            meanings = meanings.split(':')
            data.append({
                'kanji': kanji,
                'radicals': radicals,
                'meanings': meanings,
                'theme': theme,
                'subtheme': subtheme,
            })
    return data
```

## 1.2 Get List of Nodes

```python
class KanjigenNode:
    def __init__(self, name, dtype):
        self.name = name
        self.dtype = dtype
        
        mtc = {}
        mtc['kanji']   = 'red'
        mtc['radical'] = 'orange'
        mtc['meaning'] = 'meaning'
        
        self.color = mtc[dtype]

    def __repr__(self):
        return self.name + '-' + self.dtype
    
    def __str__(self):
        return self.name
```

```python
def get_list_node(data):
    list_kanji, list_radical, list_meaning, list_theme, list_subtheme = set(), set(), set(), set(), set()

    for d in data:
        list_kanji    |= set(d['kanji'])
        list_radical  |= set(d['radicals'])
        list_meaning  |= set(d['meanings'])
        list_theme.add(d['theme'])
        list_subtheme.add(d['subtheme'])

    list_radical.remove(''), list_theme.remove(''), list_subtheme.remove('')

    list_node = []
    # tuple = (0,1,2) = (symbol, type, color)
    list_node += [KanjigenNode(k, "kanji")       for k in list_kanji]
    list_node += [KanjigenNode(r, "radical")     for r in list_radical]
#     list_node += [KanjigenNode(m, "meaning")   for m in list_meaning]
#     list_node += [KanjigenNode(t, "theme")     for t in list_theme]
#     list_node += [KanjigenNode(s, "subtheme")  for s in list_subtheme]

    return list_node
```

```python
list_node = get_list_node(data)
print(list_node[:10])
```

### 1.2.1 Helper

```python
def find_node(list_node, name, dtype):
    list_node_found = list(filter(lambda node: node.name == name and node.dtype == dtype, list_node))
    if len(list_node_found) > 0:
        return list_node_found[0]
    return None
```

## 1.3 Get List of Edge

```python
def get_list_edge(data, list_node):
    list_edge = []
    
    for d in data:
        node_kanji = find_node(list_node, d['kanji'], 'kanji')
            
        for radical in d['radicals']:
            node_radical = find_node(list_node, radical, 'radical')
            if node_radical != None:
                list_edge += [(node_kanji, node_radical)]
            
#         for meaning in d['meanings']:
#             node_meaning = find_node(list_node, meaning)
#             list_edge += [(node_kanji, node_meaning)]
        

    return list_edge
```

```python
list_edge = get_list_edge(data, list_node)
print(list_edge)
```

## 2.2 Brute Force Algorithm

```python
kin = find_node(list_node, '怨', 'kanji')
kout = find_node(list_node, '浦', 'kanji')

result_has_path = nx.has_path(G, kin, kout)
result_shortest_path = nx.shortest_path(G, source=kin, target=kout)

print("has_path = ", result_has_path)
print(result_shortest_path)
```

```python
kinputs = ['怨','姻','桜']
koutputs = ['浦','奥','媛']

def find_route(kinputs, koutputs):
    
    results = []
    for kin in kinputs:
        for kout in koutputs:
            kin_node = find_node(list_node, kin, 'kanji')
            kout_node = find_node(list_node, kout, 'kanji')
            
            result_shortest_path = nx.shortest_path(G, source=kin_node, target=kout_node)
            results += [result_shortest_path]
            
    return results        

results_find_route = find_route(kinputs, koutputs)
```

```python
results_find_route
```

```python
def get_union_results_find_route(results_find_route):
    
    R_list = []
    
    
    for result in results_find_route:
        
        R = nx.Graph()
        
        result_node = []
        for r in result:
            result_node += [r]
            
        result_edge = nx.utils.pairwise(result_node)
        
        R.add_nodes_from(result_node)
        R.add_edges_from(result_edge)
        
        
        print(R.nodes())
        print(R.edges())
        print(R_list)
        print('--------------')
        R_list.append(R)
        
    return nx.compose_all(R_list)
```

```python
union_result = get_union_results_find_route(results_find_route)
```
